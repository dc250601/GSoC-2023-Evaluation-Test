```python
import numpy as np
import pyarrow.parquet as pq
import os
import time as time
import PIL.Image as imm
path = "./Images/"
import cv2
```

```python
def generate(pf):
    record_batch = pf.iter_batches(batch_size=1024)
    while(True):
        try:
            batch = next(record_batch)
            print(batch.num_rows)
            transform(batch)
        except StopIteration:
            print("Done")
            break
```

```python
alpha = 0
beta =  0
def saver(im,meta):
    global alpha
    global beta
    for i in range(meta.shape[0]):
        img = im[i,:,:,:]
        img = img.T
        img[:,:,0] = ( img[:,:,0] - img[:,:,0].min() ) / ( img[:,:,0].max()-img[:,:,0].min() )
        img[:,:,1] = ( img[:,:,1] - img[:,:,1].min() ) / ( img[:,:,1].max()-img[:,:,1].min() )
        img[:,:,2] = ( img[:,:,2] - img[:,:,2].min() ) / ( img[:,:,2].max()-img[:,:,2].min() )
        img = img*255
        img = img.astype(np.uint8)
        #lmm = imm.fromarray(img)
        if(meta[i] == 0):
            impath = path+"1/"+str(alpha)+".png"
            alpha = alpha + 1
        if(meta[i] == 1):
            impath = path+"0/"+str(beta)+".png"
            beta = beta + 1
        #lmm.save(impath)
        cv2.imwrite(impath , img)
```

```python

def transform(batch):
    p = batch.to_pandas()
    im = np.array(np.array(np.array(p.iloc[:,0].tolist()).tolist()).tolist())
    meta = np.array(p.iloc[:,3])
    saver(im,meta)
```

```python
generate(pq.ParquetFile("./Data/QCDToGGQQ_IMGjet_RH1all_jet0_run0_n36272.test.snappy.parquet"))
generate(pq.ParquetFile("./Data/QCDToGGQQ_IMGjet_RH1all_jet0_run1_n47540.test.snappy.parquet"))
generate(pq.ParquetFile("./Data/QCDToGGQQ_IMGjet_RH1all_jet0_run2_n55494.test.snappy.parquet"))
```
